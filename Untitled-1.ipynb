## Collecting data
While I have reversed engineered the game to obtain direct access to the API, it is not possible to obtain past event data through the game API as it becomes unavailable after a few days of the event finishing. However, there is an API provided by [Matsurihi.me](https://www.matsurihi.me/) that keeps past event data.

The document (Japanese) is available at https://api.matsurihi.me/docs/#sc-v1-events-fanranking. Using the API provided, we can request the past event data in JSON for analysis.

### Reading the list of past events
The code below reads the list of past events from the API and displays it for easier configuration.

```python
from dataclasses import dataclass
import requests
import time
from datetime import datetime

default_headers = {'User-Agent': 'AyyLmao'}
BASE_URI = "https://api.matsurihi.me"
VERSION_ENDPOINT = f"{BASE_URI}/sc/v1/events/fanRanking"

def get(url: str):

    response = requests.get(url, headers=default_headers);
    # Throws if code indicates non-success.
    response.raise_for_status()

    return response.json()

@dataclass
class EventDetail:
    id: int
    name: str
    begin_time: datetime
    end_time: datetime

def get_event_details() -> dict[int, EventDetail]:
    raw_response = get(VERSION_ENDPOINT)
    return {x['id']: EventDetail(x['id'], x['name'], datetime.fromisoformat(x['beginTime']), datetime.fromisoformat(x['endTime'])) for x in raw_response}


events = get_event_details()

print("List of past events:")

for id, event in events.items():
    print(f"{event.name} ({event.id}): {event.begin_time:%Y-%m-%d %H:%M} ~ {event.end_time:%Y-%m-%d %H:%M}")
```

### Reading the API data into DataFrame
The data is a list of scores at a certain time for a given rank.

From the API document, the data provided are:

* Character
* Rank
* Time
* Fan count

In addition, we calculate the following for easier analysis of the data on the last day.

* Is last day
    * There is no breaktime on the last day
    * I'm expecting people play more aggressively on the last say than other days
* Holiday
    * As people have more time on holidays, I expect people to play more
* Difference
    * For graphing
* Time elapsed since start
    * To use as x-axis when drawing graphs

The API is ratelimited - when bursts of requests are made during a short period, the API returns a 429 Too Many Requests error, asking to slow down. One simple way to avoid this is to insert a delay between requests, but given that we need to make at least 25 requests, a second delay would result in total 25 seconds of waiting (which isn't that bad, but is still long enough to be an annoyance). Given that the data is static, we can improve this by caching the data on the first time, then reading from the local cache in subsequent reads.


```python
from pathlib import Path
from pandas import DataFrame
import pandas
import json
# Configuration data

CHARACTER_IDS = map(str, range(1, 26))  # May be any type of string enumerable, such as list
EVENT_ID = 40009
RANKS = [10, 100, 1000]

# End configuration

cache_dir = Path('./cache')
cache_dir.mkdir(exist_ok=True)

@dataclass
class FanStatus:
    time: datetime
    score: int

    
@dataclass
class Participant:
    character_id: str
    rank: int
    data: list[FanStatus]

    def to_dataframe(self) -> DataFrame:
        df = DataFrame(self.data)
        df['rank'] = self.rank
        df['character_id'] = self.character_id
        df['elapsed_minutes'] = (df['time'] - events[EVENT_ID].begin_time).apply(lambda x: x.total_seconds() // 60)
        df['is_holiday'] = df['time'].apply(lambda x: x.weekday >= 5)
        df['diff'] = df['score'].diff()
        return df


def get_ranks(character_id: str) -> list[Participant]:
    # The API is ratelimited - therefore it is very expensive for us to load from the API every single time
    # So to save some time, we cache the result locally and load from the cache if possible.

    file_name = f"{EVENT_ID}_{character_id}_{','.join(map(str, RANKS))}.json"
    file_path = cache_dir / file_name
    if file_path.exists():
        response = json.loads(file_path.read_text())
    else:
        rank_endpoint = f"{BASE_URI}/sc/v1/events/fanRanking/{EVENT_ID}/rankings/logs/{character_id}/{','.join(map(str, RANKS))}"
        response = get(rank_endpoint)
        file_path.write_text(json.dumps(response))
        # The service has a ratelimit. Not sure how many requests are allowed at a given time?
        time.sleep(1)

    # Structure:
    # [
    #   {
    #     "rank" : string
    #     "data" [{"score": int, "summaryTime": iso datetime}]
    #   },
    #   data for other requested ranks...
    # }

    return [Participant(character_id, participant['rank'], [FanStatus(datetime.fromisoformat(entry['summaryTime']), entry['score']) for entry in participant['data']]) for participant in response]

ranks: list[Participant] = []

for character_id in CHARACTER_IDS:
    print(f"Reading rank data of character {character_id}");
    character_ranks = get_ranks(character_id)

    ranks += character_ranks
```

### Modifying the time provided by the API
Since the event ends at 12:00 and the final data is unavailable until 15:00, the returned data has a gap that jumps from 11:30 to 15:00.
We manipulate the time & elapsed minute columns of the last entries logged at 15:00 so they show up as 12:00 instead.

```python
from typing import Optional
from datetime import timedelta

def convert_dataframe(x: Participant):
    df = x.to_dataframe()
    max_index = df['elapsed_minutes'].idxmax()
    #max_entry = df.iloc[max_index]
    df.iloc[:, (-1, 4)] -= timedelta(minutes=210)  # type: ignore
    df.iloc[:, (-1, 0)] -= 210  # type: ignore
    return df


df: DataFrame = pandas.concat(map(lambda x: x.to_dataframe(), ranks))

maxTime = df['time'].max()
mask = df['time'] == maxTime
df.loc[mask, 'time'] -= timedelta(minutes=180)  # type: ignore
df.loc[mask, 'elapsed_minutes'] -= 180  # type: ignore
```

## Data analysis - 1. Looking at the per-group statistics

Using groupby and other methods, we take a look at various statistics.

### Looking at the final score (Using the max and rank methods)
Here, we use groupby and various methods to list and rank the final score for various types of ranks. The result is sorted the average of gold rank then silver rank, with ties ordered by the gold rank.

```python
platinum_grouped = df[df['rank'] == 10].groupby('character_id', sort=False)['score'].max()
gold_grouped = df[df['rank'] == 100].groupby('character_id', sort=False)['score'].max()
silver_grouped = df[df['rank'] == 1000].groupby('character_id', sort=False)['score'].max()
results = DataFrame([platinum_grouped, gold_grouped, silver_grouped], index=['platinum', 'gold', 'silver']).T
results_rank = results.rank(ascending=False)
results = results.join(results_rank, rsuffix="_rank")
results['popularity'] = (results['gold_rank'] + results['silver_rank']) / 2
results.sort_values(by=['popularity', 'gold_rank'])
```

Looking at the result, we can see that for 100th/1000th, the rank of the idol within the rank zone roughly matches up with the other. This suggests that the score for 100/1000th is a good indicator for popularity. Also, it is quite surprising to find that even at the same 100th rank zone the scores can differ by three-fold. This demonstrates that popularity has a huge effect on the difficulty of a particular idol up to a certain point.

However, the same cannot be said for 10th rank zone; there are many cases where the value of 'popularity' column is wildly different from the platinum rank. This suggests that for 10th and above, popularity has less of an effect on the difficulty.

### Calculating average increase in score per hour (Using the mean method)
This is a very helpful metric when deciding whether to aim for a certain rank, as you can compare the average speed with your speed to see if the rank would be viable for you or not.

```python
pandas.options.display.float_format = '{:,.2f}'.format

grouped = df.groupby(['character_id', 'rank'], sort=False)
grouped['diff'].mean() * 2  # The data has 30mins interval between values. To make it an hourly rate, we need to multiply by 2.
```

However, when interpreting this value, there is a very important caveat to consider; This is average rate over the whole of the event, meaning it does not take the mandatory 6 hours of breaktimes into account. So in order to make proper judgements, one must not compare this with their hourly rate measured for an hour directly, but instead calculate the hourly rate per day by multiplying their rate by 18 hours (hours available per day) then dividing by 24 (hours in a day).

## Data analysis - 2. Graphing
We look at the trends of the score by drawing graphs.

### Checking the impact of holidays on the hourly rate

Since people would have extra time to play the game on holidays, I expect the competition to get fiercer on holidays compared to other days. We can check this by drawing graphs of the rate of change.

```python
filtered = df[(df['character_id'] == '10') & (df['rank'] == 100)]
filtered.plot(kind='line', x='time', y='diff', c='coral')
```

It turns out, that for 100th ranks, people don't finish the produces fast enough for the difference to be continuous; in order to produce any meaningful graph we must reduce the interval to more reasonable level for this rank zone.

```python
import matplotlib.dates as mdates

# Take every fourth entry from the filtered list. (2-hour interval)
filtered_chunked: DataFrame = filtered.iloc[::4].copy()  # type: ignore
# Recalculate the difference as the previous difference was on a per-30min interval.
filtered_chunked['diff'] = filtered_chunked['score'].diff()

ax = filtered_chunked.plot(kind='line', x='time', y='diff', c='coral', figsize=(20, 5))

# Set the X ticks and format for better visibility
ax.xaxis.set_major_locator(mdates.HourLocator(interval=4))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d %H:%M'))
```

There were two holidays during the event duration; 15th and 16th, which were weekends. Looking at 10/15's data, we can see that there is a higher spike than that of 10/11~10/14, suggesting that the rate is indeed higher during the weekends than weekdays. It also is obvious that there is an even larger spike at the end of the event; this likely is a combination of both the day being a weekend and the last day of the event.

It could be argued that 10/15 having a larger spike is due to the fact that the end date is near, not only because it's weekend; however with this particular event which only had two weekend days right at the end, it is hard to isolate the cause of the spike.

(As a sidenote, it is quite obvious that people don't play the game as much at nighttime.)

## Data analysis - 3. Regression modelling

* The data is harder to model with just two variables, since we have already identified that there likely are external factors that affect the rate of change (holidays, last day, time of day etc...) apart from the time elapsed and the score
* We can try creating a regression model that takes in a normalised elapsed time and outputs the estimated score...
    * By normalizing, I mean we turn the "elapsed minute" into a number between 0 and 1, where 0 is the start of the event and 1 is the end of the event.
    * However it won't be really useful as the model would not work if the duration gets any longer

```python
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

filtered_copy = filtered.copy()
elapsed_max = filtered_copy['elapsed_minutes'].max()
filtered_copy['elapsed_ratio'] = filtered_copy['elapsed_minutes'] / elapsed_max

x = filtered_copy[['elapsed_ratio']]
y = filtered_copy['score']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)

poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
pr = LinearRegression()
pr.fit(x_train_poly, y_train)

x_test_poly = poly.fit_transform(x_test)
y_hat_test = pr.predict(x_test_poly)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.plot(x_train, y_train, 'o', label="train data")
ax.plot(x_test, y_hat_test, 'r+', label="predicted value")
ax.legend(loc='best')
plt.show()
plt.close()
```
